```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm; tqdm.pandas()
pd.options.display.max_columns = 202
pd.options.display.max_rows = 300
```

```python
train = pd.read_pickle('data/train_original.pkl')
test = pd.read_pickle('data/test_original.pkl')
special_cols = [col for col in train.columns if train[col].dtype != np.float64]
feature_cols = [col for col in train.columns if col not in special_cols]
```

```python
for i, col in enumerate(tqdm(feature_cols)):
    fut_values = np.concatenate(pd.concat([train[feature_cols[i+1:]], test[feature_cols[i+1:]]], axis=0).values)
    fut_vc = pd.Series(fut_values).value_counts()
    train[col] = train[col].map(fut_vc)
    test[col] = test[col].map(fut_vc)

train.head(3)
```

```python
train = train[feature_cols]
test = test[feature_cols]
```

```python
feature_cols = [f+'_future_appear_cnts' for f in feature_cols]
train.columns=feature_cols
test.columns=feature_cols
```

```python
train.to_pickle('features/future_appear_count_train.pkl')
test.to_pickle('features/future_appear_count_test.pkl')
```
