```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm; tqdm.pandas()
pd.options.display.max_columns = 202
pd.options.display.max_rows = 300
```

```python
train = pd.read_pickle('data/train_original.pkl')
test = pd.read_pickle('data/test_original.pkl')
special_cols = [col for col in train.columns if train[col].dtype != np.float64]
feature_cols = [col for col in train.columns if col not in special_cols]
```

```python
trn_vc = pd.Series(np.concatenate(train[feature_cols].values)).value_counts()
tst_vc = pd.Series(np.concatenate(test[feature_cols].values)).value_counts()
all_df = pd.concat([train[feature_cols], test[feature_cols]], axis=0).reset_index(drop=True)
all_vc = pd.Series(np.concatenate(all_df.values)).value_counts()
```

```python
th=500
pop_values = all_vc[all_vc>th].index
(all_vc>th).sum()
```

```python
best_col_info = []

for i, val in enumerate(pop_values):
    per_col_cnts = [(all_df[f]==val).sum() for f in feature_cols]
    best_col = feature_cols[per_col_cnts.index(max(per_col_cnts))]
    print('{} {} best col {}'.format(i, val, best_col))
    best_col_info.append(per_col_cnts.index(max(per_col_cnts)))
    
best_col_info = pd.Series(index=pop_values, data=best_col_info)
```

```python
train['pop_val_cnt_per_row'] = train[feature_cols].progress_apply(lambda row: row.isin(pop_values).sum(),axis=1)
test['pop_val_cnt_per_row'] = test[feature_cols].progress_apply(lambda row: row.isin(pop_values).sum(),axis=1)
```

```python
def analyze_corr(vals):
    tmp = pd.DataFrame()
    tmp['target'] = train.target
    tmp['val'] = vals
    return tmp.corr('spearman')
```

```python
analyze_corr(train['pop_val_cnt_per_row'])
```

```python
def get_min_distance(row):
    filt = row.isin(pop_values)
    if filt.sum() == 0:
        return -999
    else:
        min_pos = np.arange(len(feature_cols))[filt][0]
        val = row[row.isin(pop_values)].values[0]
        return min_pos-best_col_info.loc[val]
        
train['pop_val_min_dist'] = train[feature_cols].progress_apply(get_min_distance, axis=1)
test['pop_val_min_dist'] = test[feature_cols].progress_apply(get_min_distance, axis=1)
analyze_corr(train['pop_val_min_dist'])
```

```python
def get_max_distance(row):
    filt = row.isin(pop_values)
    if filt.sum() == 0:
        return -999
    else:
        max_pos = np.arange(len(feature_cols))[filt][-1]
        val = row[row.isin(pop_values)].values[-1]
        return max_pos-best_col_info.loc[val]
    
train['pop_val_max_dist'] = train[feature_cols].progress_apply(get_max_distance, axis=1)
test['pop_val_max_dist'] = test[feature_cols].progress_apply(get_max_distance, axis=1)
analyze_corr(train['pop_val_max_dist'])
```

```python
train[['pop_val_cnt_per_row']].to_pickle('features/pop_val_cnt_per_row_train')
test[['pop_val_cnt_per_row']].to_pickle('features/pop_val_cnt_per_row_test')
```
