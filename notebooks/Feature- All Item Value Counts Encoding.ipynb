```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm; tqdm.pandas()
pd.options.display.max_columns = 202
pd.options.display.max_rows = 300
```

```python
train = pd.read_pickle('data/train_original.pkl')
test = pd.read_pickle('data/test_original.pkl')
special_cols = [col for col in train.columns if train[col].dtype != np.float64]
feature_cols = [col for col in train.columns if col not in special_cols]
```

```python
all_values = pd.concat([train[feature_cols], test[feature_cols]], axis=0).values.reshape((-1))
all_vc = pd.Series(all_values).value_counts()
```

```python
for col in tqdm(feature_cols):
    train[col] = train[col].map(all_vc)
    test[col] = test[col].map(all_vc)
```

```python
train.head(3)
```

```python
for df in [train, test]:
    df['vc_encode_max'] = df[feature_cols].max(axis=1)
    df['vc_encode_min'] = df[feature_cols].min(axis=1)
    df['vc_encode_sum'] = df[feature_cols].sum(axis=1)
    df['vc_encode_std'] = df[feature_cols].std(axis=1)
```

```python
final_feat_cols = ['vc_encode_max', 'vc_encode_min', 'vc_encode_sum', 'vc_encode_std']
```

```python
train[['target']+final_feat_cols].corr('spearman')
```

```python
train[feature_cols].to_pickle('features/all_items_vc_train_v1.pkl')
test[feature_cols].to_pickle('features/all_items_vc_test_v1.pkl')
```
