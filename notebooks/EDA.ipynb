```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm; tqdm.pandas()
pd.options.display.max_columns = 202
pd.options.display.max_rows = 300
```

```python
train = pd.read_pickle('data/train_original.pkl')
test = pd.read_pickle('data/test_original.pkl')
```

```python
train.head(3)
```

```python
special_cols = [col for col in train.columns if train[col].dtype != np.float64]
special_cols
```

```python
feature_cols = [col for col in train.columns if col not in special_cols]
```

```python
trn_vc = pd.Series(np.concatenate(train[feature_cols].values)).value_counts()
tst_vc = pd.Series(np.concatenate(test[feature_cols].values)).value_counts()
all_df = pd.concat([train[feature_cols], test[feature_cols]], axis=0).reset_index(drop=True)
all_vc = pd.Series(np.concatenate(all_df.values)).value_counts()
```

```python
def plot_target_val_loc(target_val):
    filt = train[feature_cols].progress_apply(lambda row: target_val in row.values ,axis=1)
    possible_cols = [c for c in feature_cols if target_val in train[c].values]
    
    fig = plt.figure(figsize=(20, 200))
    #fig.subplots_adjust(hspace=0.4, wspace=0.4)
    print(filt.sum())
    bitmap = train.loc[filt, possible_cols+['target']].values
    bitmap = (bitmap==target_val) | (bitmap==1)
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.imshow(bitmap)
```

```python
selected_vals = trn_vc[(trn_vc>th)].index.tolist()
res = pd.DataFrame(columns=['pop val', 'pop val col', 'pop val cnt'])

for i, val in tqdm(enumerate(selected_vals)):
    per_col_cnts = [(train[f]==val).sum() for f in feature_cols]
    best_col = feature_cols[per_col_cnts.index(max(per_col_cnts))]
    res.loc[i] = [val, best_col, trn_vc[val]]
    
res['pop val col'].nunique()
```

```python
selected_vals = tst_vc[(tst_vc>th)].index.tolist()
res = pd.DataFrame(columns=['pop val', 'pop val col', 'pop val cnt'])

for i, val in tqdm(enumerate(selected_vals)):
    per_col_cnts = [(test[f]==val).sum() for f in feature_cols]
    best_col = feature_cols[per_col_cnts.index(max(per_col_cnts))]
    res.loc[i] = [val, best_col, tst_vc[val]]
    
res['pop val col'].nunique()
```

```python
def plot_target_val_list_loc(target_val_list):
    filt = None 
    possible_cols = None 
    
    for tv in tqdm(target_val_list):
        tmp_filt = train[feature_cols].apply(lambda row: tv in row.values ,axis=1)
        tmp_p_cols = [c for c in feature_cols if tv in train[c].values]
        
        if filt is None:
            filt = tmp_filt
            possible_cols = tmp_p_cols
        else:
            filt = filt | tmp_filt
            possible_cols += tmp_p_cols
            
            
    possible_cols = list(set(possible_cols))
    print(filt.sum(), len(possible_cols))
    
    fig = plt.figure(figsize=(20, 500))
    #fig.subplots_adjust(hspace=0.4, wspace=0.4)
    
    values = train.loc[filt, possible_cols+['target']].values
    all_filt = None
    for tv in tqdm(target_val_list):
        if all_filt is None:
            all_filt = (values==tv) | (values==1)
        else:
            all_filt = all_filt | (values==tv)
        
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.imshow(all_filt)
```

```python
plot_target_val_list_loc(trn_vc[(trn_vc>th)].index)
```

```python
th=500
pop_values = all_vc[all_vc>th].index
(all_vc>th).sum()
```

```python
plt.hist(pop_values); plt.show()
```

```python
best_col_info = []

for i, val in enumerate(pop_values):
    per_col_cnts = [(all_df[f]==val).sum() for f in feature_cols]
    best_col = feature_cols[per_col_cnts.index(max(per_col_cnts))]
    print('{} {} best col {}'.format(i, val, best_col))
    best_col_info.append(per_col_cnts.index(max(per_col_cnts)))
    
best_col_info = pd.Series(index=pop_values, data=best_col_info)
```

```python
#th = 300
#pop_values = trn_vc[(trn_vc>th)].index.tolist()
#print((trn_vc>th).sum(), trn_vc[(trn_vc>th)].sum())
#pop_values = trn_vc.index[trn_vc.index.isin(all_vc.index[:237])].tolist()

pop_val_cnt_per_row = train[feature_cols].progress_apply(lambda row: row.isin(pop_values).sum(),axis=1)
```

```python
pop_val_cnt_per_row.value_counts()
```

```python
def analyze_corr(vals):
    tmp = pd.DataFrame()
    tmp['target'] = train.target
    tmp['val'] = vals
    return tmp.corr('spearman')
```

```python
analyze_corr(pop_val_cnt_per_row)
```

```python
def get_min_distance(row):
    filt = row.isin(pop_values)
    if filt.sum() == 0:
        return -999
    else:
        min_pos = np.arange(len(feature_cols))[filt][0]
        val = row[row.isin(pop_values)].values[0]
        return min_pos-best_col_info.loc[val]
        
dist = train[feature_cols].progress_apply(get_min_distance, axis=1)
plt.hist(dist[dist!=-999]); plt.show()
analyze_corr(dist)
```

```python
def get_max_distance(row):
    filt = row.isin(pop_values)
    if filt.sum() == 0:
        return -999
    else:
        max_pos = np.arange(len(feature_cols))[filt][-1]
        val = row[row.isin(pop_values)].values[-1]
        return max_pos-best_col_info.loc[val]
    
dist = train[feature_cols].progress_apply(get_max_distance, axis=1)
plt.hist(dist[dist!=-999]); plt.show()
analyze_corr(dist)
```
