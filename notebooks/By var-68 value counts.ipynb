```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm; tqdm.pandas()
pd.options.display.max_columns = 202
pd.options.display.max_rows = 300
```

```python
train = pd.read_pickle('data/train_original.pkl')
test = pd.read_pickle('data/test_original.pkl')
special_cols = [col for col in train.columns if train[col].dtype != np.float64]
feature_cols = [col for col in train.columns if col not in special_cols]
```

```python
all_df = pd.concat([train[feature_cols], test[feature_cols]], axis=0)
```

```python
all_df['var_68'].nunique()
```

```python
for col in tqdm(feature_cols):
    if col == 'var_68': 
        continue
        
    for date in all_df['var_68'].unique():
        filt = all_df.var_68 == date
        all_df.loc[filt, col+'_by_date_freq'] = all_df.loc[filt, col].map(all_df.loc[filt, col].value_counts())
```

```python
selected_features = [c for c in all_df.columns if '_by_date_freq' in c]
all_df.iloc[:train.shape[0], ][selected_features].to_pickle('features/by_date_freq_train.pkl')
all_df.iloc[train.shape[0]:, ][selected_features].reset_index(drop=True).to_pickle('features/by_date_freq_test.pkl')
```
